```python
#here i converting the '0' into integer :
df['Class'] = df['Class'].astype(str).str.replace("'", "").astype(int)
```

```python
#placing the legit & fraud  value in differentr variable :
legit=df[df.Class==0]
fraud=df[df.Class==1]

print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures
print(legit.Amount.describe())
print('*'*20)
print('*'*20)
print(fraud.Amount.describe())
```

```python
# compair the values here:\
df.groupby('Class').mean()
```

```python
# deling the data using undersampling here :

legit_sample=legit.sample(n=492)

new_df=pd.concat([legit_sample,fraud],axis=0)
```

```python
print(new_df.shape)
```

```python
#here new dataset head:
new_df.head()
```

```python
new_df['Class'].value_counts()
```

```python
# new statsitical measure:
new_df.groupby('Class').mean()
```

```python
# spliting the data from here:
from sklearn.model_selection import train_test_split
```

```python
X=new_df.drop(columns='Class',axis=1)
Y=new_df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20,stratify=Y, random_state=42)
print(X.shape,X_train.shape,X_test.shape)


from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# model traing here:
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
model={
    'LogisticRegression':LogisticRegression(),
    'RandomForestClassifier':RandomForestClassifier(n_estimators=100,
    max_depth=10)
}

from sklearn.metrics import classification_report,accuracy_score
for name,m in model.items():
  print(f"model:{name}")
  m.fit(X_train,y_train)
  y_pred=m.predict(X_test)
  print(classification_report(y_test,y_pred))
  print("*"*50)

```

```python
# testing my model on random data :
input_data= X_test[0]

#convert into array:
input_data=np.array(input_data).reshape(1,-1)

#scaling if it needed :
input_data=scaler.transform(input_data)

try:
    input_data = scaler.transform(input_data)
except:
    pass

# Select a specific model from the dictionary for prediction
# For example, using the RandomForestClassifier:
prediction = model['RandomForestClassifier'].predict(input_data)

if prediction[0]==0:
  print('valid transaction')
else:
  print('invalid transaction')
```

```python
# saving the model here:
import joblib
joblib.dump(model['RandomForestClassifier'],'credit_card_model.pkl')
joblib.dump(scaler,'scaler.pkl')
joblib.dump(X.columns.tolist(),'columns.pkl')
```
